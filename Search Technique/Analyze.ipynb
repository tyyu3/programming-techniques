```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("timings.txt",header=None)
df = df.iloc[:, 1:]
df
```

```python
vals = df.values.tolist()
x = vals[0]
y1 = [x/10**9 for x in vals[1]]
y2 = [x/10**9 for x in vals[2]]
y3 = [x/10**9 for x in vals[3]]
y4 = [x/10**9 for x in vals[4]]
```

```python
plt.figure(figsize=(8,8))
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.xlabel('N', fontsize=18)
plt.ylabel('Time, seconds', fontsize=16)
plt.legend(['Linear Search','Multimap','Presorted Binary Search','Sort & Binary Search'])
plt.savefig("Search_w_sort.png")
```
